# Philippine Fuel Price Volatility Analysis

## Data Source
* **Source**: Department of Energy (DOE) Philippines
    - A weekly report from the department of energy captures ranges of fuel prices from different fuel products. The report includes various fuel products, with RON95 and Diesel selected to be used for the analysis. This report covers the needed data for the research question.
    - Choosing the DOE as the data source due to its credibility being a government organization website. With the source specifically focuses on energy trends in the Philippines, fuel and oil being a main focus of such organization
 * **Rejecting Other Sources**
      - Other sources were also evaluated as a data source such as GlobalPetrolPrices, CEIC, and Facebook Page Fuel Updates. These sources were rejected such as having to pay for a data set, inconsistencies, and lacking of credibility.
 * **Limitation of Data Source**
    - Small Inconsistencies
        - The data source has inconsistency with its monitoring with some lasting a week, while others just lasting a day or two. This inconsistency is what drove the decision of having a monthly timeline of data collection instead of each monitoring document directly from the source
    - Formatting
        - The formatting was limiting the data collection with it being a PDF and thus having to convert it into an Excel file to properly collect the average of the needed product value correctly.
    - Coverage Limitation
        - As the data source only reports up until late of December 2020, this limits the data source for much historical data regarding fuel prices.
* **Coverage**: 2022 to present
    - As the weekly became more consistent in 2022, thus becoming the starting point of the coverage of the dataset.
    - A 5 years comparison would be the best for this analysis for the reason of wanting consistency and relevance.
    - Year 2022 was also chosen rather than much later years due to wanting to find out relationships of the changes in fuel prices in correlation to major events.

## Data Collection Notes
* **Methodology Decisions**
    - Timeline
        - As the datasource has inconsitencies with monitoring such that some are weekly monitoring and other having just 2 day monitoring for each PDF for that month. Deciding that getting data per month will be the best decision for this dataset.
    - Fuel Products
        - As the most common fuel used by Filipinos are RON95 and Diesel fuels, these products are what selected to be used for this dataset.
    - Values
        - With the timeline being monthly, getting the mean per month for each monitoring PDF will be done for this dataset. As each PDF has NCR areas with different range values, averaging each PDF will be done as well, afterward the result of all computation is the value for each fuel products.

* **Column Structure**
    - Year & Month
        - Having a separate column for the year and month is done as to have a better comparison for analysing changes / differences when it comes to yearly fuel price changes.
    - Ron95_avg
        - In between RON91 and RON100, one of the most commonly used fuel in the country, RON95 is chosen as the product of this analysis. This product is also chosen as to have a better comparison with Diesel as RON95 being used in most private vehicles as such the volatility is important to everyday Filipino private vehicle owners.
    - Diesel_avg
        - Being one of the most common fuel product in the country used by industrial vehicles, Diesel is chosen as the product of this analysis. Being used in industrial vehicles, the cost of transporting goods across the country is directly affected by the volatility of this product.
